```python
import cv2
import keras
import numpy as np
from google.colab.patches import cv2_imshow

from google.colab import drive 
drive.mount("/content/drive")
```

```python
lena = cv2.imread("/content/drive/MyDrive/PCD/lena.jpeg")
```

```python
print("lena")
cv2_imshow(lena)
```

```python
lena_grey = (cv2.cvtColor(lena, cv2.COLOR_BGR2GRAY))
print("lena Abu")
cv2_imshow(lena_grey)
```

```python
def gauss_noise(image, gauss_var=1000):
    mean = 0
    sigma = gauss_var ** 0.5
    gauss = np.random.normal(mean, sigma, image.shape)

    res = image + gauss
    noisy = np.clip(res, 0, 255).astype(np.uint8)
    return noisy
```

```python
lena_noisy = gauss_noise(lena_grey, gauss_var=25)
print("lena Abu Noisy")
cv2_imshow(lena_noisy)
```

```python
lena_colored_noisy = gauss_noise(lena, gauss_var=25)
print("lena Colored Noisy")
cv2_imshow(lena_colored_noisy)
```
