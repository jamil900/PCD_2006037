Import 

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import drive 
drive.mount("/content/drive")
from matplotlib import pyplot as plt
```

Mengambil Gambar

```python
ayam = cv2.imread("/content/drive/MyDrive/PCD/animals-birds-wallpaper-preview.jpg")
```

Menampilkan Gambar

```python
print("ayam")
cv2_imshow(ayam)
```

Menampilkan Dimensi

```python
print("anak ayam")
print(ayam.shape)
```

Menampilkan Konversi Warna

```python
ayam_abu = (cv2.cvtColor(ayam, cv2.COLOR_BGR2GRAY))
ayam_rgb = (cv2.cvtColor(ayam, cv2.COLOR_BGR2RGB))
ayam_lab = (cv2.cvtColor(ayam, cv2.COLOR_BGR2LAB))
ayam_hsv = (cv2.cvtColor(ayam, cv2.COLOR_BGR2HSV))

print("anak ayam warna abu")
cv2_imshow(ayam_abu)
print()

print("anak ayam warna rgb")
cv2_imshow(ayam_rgb)
print()

print("anak ayam warna lab")
cv2_imshow(ayam_lab)
print()

print("anak ayam warna hsv")
cv2_imshow(ayam_hsv)
```

Menggunakan MatPolib

```python
plt.figure(figsize=[15,15])
plt.subplot(221)
plt.imshow(ayam_abu)
plt.subplot(222)
plt.imshow(ayam_rgb)
plt.subplot(223)
plt.imshow(ayam_lab)
plt.subplot(224)
plt.imshow(ayam_hsv)
plt.show()
```
